```python
def sum_32 (string):
  x=[]
  for i in range(0,62,2):
    s=0
    for j in range(0,4):
      if(string[i+j]=="1"):
        s+=1
    s=float(s/4)
    x.append(s)


    #l=np.array(l)
  return x
```

```python
df["Bits_32"]=df["Bitstream"].apply(sum_32)
```

```python
df["Bits_32"][0]
```

```python
def sum_16 (string):
  x=[]
  for i in range(0,16):
    s=0
    for j in range(0,4):
      if(string[4*i+j]=="1"):
        s+=1
    s=float(s/4)
    x.append(s)


    #l=np.array(l)
  return x
```

```python
df["Bits_sum"]=df["Bitstream"].apply(sum_16)
```

```python
# Convert the column into a NumPy array with data type float32
data_array1 = np.array(df['Bits_32'].apply(lambda x: np.array(x, dtype=np.float32).tolist()).to_list(), dtype=np.float32)
data_array2 = np.array(df['Bits_sum'].apply(lambda x: np.array(x, dtype=np.float32).tolist()).to_list(), dtype=np.float32)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_array1, y, test_size=0.2, random_state=42)

```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Create a Sequential model
model2 = keras.Sequential()

# Add an input layer with 31 input features
model2.add(layers.Input(shape=(31,)))


model2.add(layers.Dense(128, activation='relu'))
model2.add(layers.Dense(64, activation='relu'))
model2.add(layers.Dense(32, activation='relu'))

# dropout layer for regularization
#model.add(layers.Dropout(0.2))

# Add more hidden layers
model2.add(layers.Dense(32, activation='relu'))
model2.add(layers.Dense(16, activation='relu'))

# Add the output layer with a single neuron and a sigmoid activation function for binary classification
model2.add(layers.Dense(1, activation='sigmoid'))

# Compile the model
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print a summary of the model architecture
model2.summary()
```

```python
# Train (fit) the model2 on the training data
model2.fit(x_train, y_train, epochs=10, batch_size=20, validation_data=(x_test, y_test))

```

```python
from sklearn.model_selection import train_test_split
s_train, s_test, y_train, y_test = train_test_split(s, y, test_size=0.2, random_state=42)

```

```python


# Create a Sequential model
model = keras.Sequential()

# Add an input layer with the shape of your sequences
model.add(layers.Input(shape=(31, 1)))  # Input shape adjusted for your data

# Add an LSTM layer with 64 units
model.add(layers.LSTM(64, return_sequences=True))  # Adjust return_sequences as needed

# Add a second LSTM layer for deeper representation
model.add(layers.LSTM(64, return_sequences=False))

# Add a dropout layer for regularization
#model.add(layers.Dropout(0.2))

# Add one or more dense layers for classification
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))

# Add the output layer with a single neuron and a sigmoid activation function for binary classification
model.add(layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print a summary of the model architecture
model.summary()
```

```python
# Train (fit) the model on the training data
model.fit(s_train, y_train, epochs=10, batch_size=25, validation_data=(s_test, y_test))

```
